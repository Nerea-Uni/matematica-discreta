**Ejercicio 1.** ¿Puede  haber  un grafo de 5 vértices cuyos  grados  sean 5,4,4,3,2? ¿Puede haber un grafo de 6 vértices cuyos grados sean 5,5,4,3,3,2?

```python
#Como se puede leer en el enunciado de la práctica, durante todos los ejercicios vamos 
#a considerar que 'grafo' significa 'grafo simple'. Un grafo simple es aquel que no
#contiene bucles ni aristas paralelas.

#Dicho esto, pasamos a analizar el primer apartado de este ejercicio.
#    Al ser un grafo simple de 5 vértices, es imposible que el primer vértice tenga 
#  grado 5, porque no puede tener bucles ni aristas paralelas y, por tanto, no habría
#  tantos vértices en dicho grafo como para cumplir ambas condiciones (que tenga 
#  grado 5 y que no contenga bucles ni aristas paralelas). Por tanto, la respuesta
#  al primer apartado sería negativa.

#Ahora analizaremos el segundo apartado:
#    En este caso no surge el problema del apartado anterior, porque todos los grados
#  de los vértices son menores que el número total de estos. En este caso sí que 
#  existe dicho grafo y sería el siguiente:
G1=Graph([(1,2),(1,3),(1,4),(1,5),(1,6),(2,3),(2,4),(2,5),(2,6),(3,4),(3,5)])
show(G1)
```

```python
#Ahora vamos a comprobar que los grados del grafo que hemos construido coinciden con
#los que nos solicita el enunciado:
G1.degree()
#Como podemos observar, sí que coinciden, por lo que la respuesta a este segundo 
#apartado es afirmativa.
```

**Ejercicio 2.** ¿Puede haber un grafo de 6 vértices en el que todos tengan grado 3?

```python
#Ahora vamos a comprobar que los grados del grafo que hemos construido coinciden con
#Para hallar el grafo que cumple dichas condiciones podemos unir dos conjuntos de 3
#vértices cada uno tal que así:
#    1 ·          · 4
#    2 ·          · 5
#    3 ·          · 6
#Entonces unimos sus vértices de la siguiente manera, consiguiendo que todos ellos 
#tengan grado 3.
G2=Graph([(1,4),(1,5),(1,6),(2,4),(2,5),(2,6),(3,4),(3,5),(3,6)])
show(G2)
```

```python
#Ahora comprobamos que el grado de todos los vértices es 3. Como observamos que sí que
#lo es, el ejercicio estaría resuelto.
G2.degree()
```

**Ejercicio 6.** Consideremos el grafo de vértices 1, 2, 3, 4, 5 y 6 y aristas (1,2),(1,4),(1,5),(2,3),(2,5),(2,6),(3,5),(3,6),(4,5),(5,6). Hallar el número de caminos de longitud 3 que unen el vértice 1 y el 6 y describir éstos.

```python
#Para llegar a la solución de este ejercicio debemos hallar la matriz de adyacencia y
#luego elevarla a 3. Una vez hemos hecho esto, miramos el elemento m_16.
G6=Graph([(1,2),(1,4),(1,5),(2,3),(2,5),(2,6),(3,5),(3,6),(4,5),(5,6)])
show(G6)
M=G6.adjacency_matrix()
show(M)
```

```python
sol6=M^3
show(sol6)


#Como podemos observar, el elemento m_16 es 5. Esto quiere decir que existen 5 caminos
#de longitud 3 que unen los vértices 1 y 6. Dichos caminos son los siguientes:
#   1·2·3·6, 1·2·5·6, 1·4·5·6, 1·5·2·6, 1·5·3·6
```

**Ejercicio 7**  Sea G el grafo de vértices 1,2,3,4,5 y 6 y aristas (1,2),(1,3),(1,4),(2,3),(2,5),(2,6),(3,4)y(3,6). Comprobar que no es un árbol y suprimir aristas para obtener un árbol generador de G.

```python
#Tal y como hemos hecho en alguno de los ejercicios anteriores, creamos un grafo con
#las características indicadas en el enunciado.
G7=Graph([(1,2),(1,3),(1,4),(2,3),(2,5),(2,6),(3,4),(3,6)])
show(G7)
```

```python
#Usamos la función .is_tree() para comprobar si es o no un árbol.
G7.is_tree()
```

```python
#Al no ser un árbol, vamos a suprimir las aristas que forman bucles, impidiendo que el
#grafo sea un árbol. Dichas aristas son, por ejemplo, (1,3), (1,4) y (2,3). Así pues, 
#así es como quedaría el grafo:
G7.delete_edges([(1,3),(1,4),(2,3)])
show(G7)
```

```python
#Comprobamos que ahora el grafo G7 sí que es un árbol.
G7.is_tree()
```

**Ejercicio 9.** Sea G el grafo de vértices 1,2,3,4,5,6 y 7 y aristas con peso (1,2,4),(1,3,6),(1,4,5),(1,6,8),(1,7,7),(2,3,3),(2,5,4),(2,7,8),(3,4,5),(3,5,6),(3,6,9),(4,5,8),(4,6,3),(5,6,9),(5,7,4) y (6,7,6). Hallar, por el algoritmo de Kruskal, un árbol generador mínimo de G.

```python
#Creamos el grafo ponderado del enunciado de la siguiente manera:
G9=Graph({1:{2:4,3:6,4:5,6:8,7:7},2:{3:3,5:4,7:8},3:{4:5,5:6,6:9},4:{5:8,6:3},5:{6:9,7:4},6:{7:6}})
G9.plot(edge_labels=True)
```

```python
#Ahora debemos convertir las etiquetas en pesos mediante la siguiente instrucción, para 
#más tarde implementar el algoritmo de Kruskal
from sage.graphs.spanning_tree import kruskal

G9.weighted(True)
kruskal(G9,check=True)
```
